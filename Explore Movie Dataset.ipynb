**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
# 添加净利润列
movie_data['profits'] = movie_data['revenue'] - movie_data['budget']
movie_data['profits_adj'] = movie_data['revenue_adj'] - movie_data['budget_adj']

# 设置年份分箱
year_width = 1
year_edges = np.arange(movie_data['release_year'].min(),movie_data['release_year'].max()+year_width,year_width)
year_center = (year_edges + year_width/2)[:-1]

# 按照分箱设置分割,设为左闭合区间
year_cut = pd.cut(movie_data['release_year'], year_edges, include_lowest = True)

# 计算分箱中的利润中心值
profits_mean = movie_data['profits'].groupby(year_cut).mean()
profits_adj_mean = movie_data['profits_adj'].groupby(year_cut).mean()

# 绘制折线图
plt.figure(figsize = [15,6])
# 未经通胀调整的票房利润
plt.subplot(1,2,1)
plt.errorbar(x = year_center, y = profits_mean);
plt.xlabel('release_year');
plt.ylabel('profits');
plt.title('profits vary by year without inflation');
# 经通胀调整的票房利润
plt.subplot(1,2,2)
plt.errorbar(x = year_center, y = profits_adj_mean);
plt.xlabel('release_year');
plt.ylabel('profits_adj');
plt.title('profits vary by year with inflation');
```

从上图可以看出，随着年份的增加，未经通胀调整的整体市场票房净利润基本呈上升趋势；而经过通胀调整计算后，净利润呈现下降的趋势

**[选做]任务3.3：**选择最多产的10位导演（电影数量最多的），绘制他们排行前3的三部电影的票房情况，并简要进行分析。

```python
# 需要忽略 导演名为miss的列
top10_index = movie_data[movie_data['director'] != 'miss'].director.value_counts()[:10].index
# 获得包含这些导演的数据行，并按照票房排序
top10_director = movie_data[movie_data['director'].isin(np.array(top10_index))].sort_values(by = 'revenue', ascending = False)
# 按导演进行分组，并获取前三部
top10_director_3 = top10_director.groupby(['director']).head(3).sort_values(by = 'director')

# 绘制簇状柱形图
plt.figure(figsize = [10,15])
sb.barplot(data = top10_director_3, y = 'original_title', x = 'revenue', hue = 'director',dodge=False);
```

从图上可知，在这些电影中，Steven Spielberg 的总体票房表现是最好的，而Woody Allen票房最差，也比较稳定。Ron Howard 和 Brian De Palma 的电影有个别电影要比他们的其他电影票房高很多。大部分导演的电影在8亿以下。

**[选做]任务3.4：**分析1968年~2015年六月电影的数量的变化。

```python
june_movie = movie_data[movie_data['release_date'].str.contains(r'^6/')].query('release_year >= 1968 & release_year <= 2015')
plt.figure(figsize = [10,12])
sb.countplot(data = june_movie, y = 'release_year', color = base_color);
```

可以看出，6月份的电影数量随着年份的增长有着明显的上升趋势，而且从2000年之后，这种趋势更加明显

**[选做]任务3.5：**分析1968年~2015年六月电影 `Comedy` 和 `Drama` 两类电影的数量的变化。

```python
plt.figure(figsize = [18,6])
for genres in ['Comedy','Drama']:
    genres_df = june_movie[june_movie['genres'].str.contains(genres)]
    max_year = genres_df['release_year'].max()
    min_year = genres_df['release_year'].min()
    year_edges = np.arange(min_year-1,max_year+1,1)
    year_center = (year_edges + 0.5)[:-1]
    year_cut = pd.cut(genres_df['release_year'], year_edges, include_lowest = True)
    counts = genres_df.genres.groupby(year_cut).count()
    plt.errorbar(x = year_center, y = counts, label = genres)
plt.legend(fontsize = 'large');
plt.grid();
```
